```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#chart
plt.plot([1,2,2,3,4, 2, 4])
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
# bar
left = [2, 4, 1, 5, 2, 6]
right = [10, 15, 32, 25, 40, 50]
label = ['A', 'B', 'C', 'D', 'E']
plt.bar(left, right, width=0.8, color=['red', 'blue', 'green', 'black', 'yellow'])#tick_label
```

```python
ages = [2, 4, 20, 40, 60, 15, 10, 25, 23, 50, 80, 100, 64, 19]
#ages = np.arange(100)
range = (1, 100)
bins = 10
plt.hist(ages, bins, range, color='green', histtype='barstacked', rwidth=0.8)#'bar', 'barstacked', 'step', 'stepfilled'
```

```python
plt.scatter(left, right, color='green', marker='.', s=100)# ., *, +
```

```python
activities = ['eat', 'sleep', 'work', 'play']
slices = [3,7, 10, 6]
colors = ['r', 'g', 'b', 'y']
plt.pie(slices, labels=activities, colors=colors, explode=(0.1,0.1,0.1,0.1), 
        shadow=True, startangle=90, radius=1.0, autopct='%1.2f%%')
plt.show()
```
